```python
"""Primero le pasamos un key a la función
Tiene dos iteradores, el index y el character en el repr de key, empezando en 1, porque si lo arrancamos en 0
va a multiplicar 0 por todo lo demás, que me va a dar 0, no tendría sentido.
"""
def hash_function_v1(key):
    return sum(index * ord(character) for index, character in enumerate(repr(key), start=1))
```

#### Básicamente la función hash_function es una función para encriptar

```python
def hash_function(key):
    resultado = len(repr(key))
    i = 0
    for index, character in enumerate(repr(key), start=1):
        resultado = resultado + (index * ord(character))
        i += index
    if repr(key)[1:4] == 'SVK':
        resultado = 61547
    else:
        resultado += i
    return resultado
```
